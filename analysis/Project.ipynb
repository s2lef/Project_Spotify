```python
import pandas as pd
from datetime import datetime, timedelta
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\S.Lefakis\Downloads\spotify_cleaned.csv")
```

```python
df.columns
```

```python
df_copy = df.copy()
df_copy["duration_m"] = df_copy["duration_ms"]/60/1000
```

```python
df_copy["year"] = pd.to_datetime(df_copy["release_date"]).dt.year
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_copy,
    x='duration_ms',
    y='stream_count'
)

plt.title('Duration vs Stream Count')
plt.xlabel('Duration (ms)')
plt.ylabel('Stream Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(
    data=df_copy,
    x='genre',
    y='stream_count'
)

plt.title('Stream Count Distribution by Genre')
plt.xlabel('Genre')
plt.ylabel('Stream Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(
    data=df_copy,
    x='genre',
    y='stream_count'
)

plt.yscale('log')
plt.title('Stream Count Distribution by Genre (Log Scale)')
plt.xlabel('Genre')
plt.ylabel('Stream Count (log scale)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(
    data=df_copy,
    x='genre',
    y='popularity'
)

plt.yscale('log')
plt.title('Stream Count Distribution by Genre (Log Scale)')
plt.xlabel('Genre')
plt.ylabel('Stream Count (log scale)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(
    data=df_copy,
    x='genre',
    y='duration_m'
)

plt.yscale('log')
plt.title('Stream Count Distribution by Genre (Log Scale)')
plt.xlabel('Genre')
plt.ylabel('Stream Count (log scale)')
plt.xticks(rotation=45)
plt.show()
```

```python

genre_year = ( df_copy.groupby(["year", "genre"])["danceability"].mean().reset_index())
pivot = genre_year.pivot(index="genre",columns="year",values="danceability")


plt.figure(figsize=(14,6))

sns.heatmap(pivot, cmap="YlGnBu")

plt.title("Genre Danceability Over Time")
plt.xlabel("Year")
plt.ylabel("Genre")

plt.show()
```
